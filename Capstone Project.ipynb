## Task 2
#### The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. 
#### Replace the missing values by running average. You should use the “For” loop for this task.

```python
project_dataframe = pd.read_csv("project.csv") #reading the project.csv file and loading its contents into the dataframe

# Initializing counters for calculating the average Cost
count = 0  # Number of non-missing Cost values
total = 0  # Sum of non-missing Cost values

# looping through each row to calculate total and count of valid cost values
for i in range(len(project_dataframe)):
    
    if not pd.isna(project_dataframe.loc[i,"Cost"]): # checking if the cost value is not missing (NaN)
        count += 1
        total += project_dataframe.loc[i,"Cost"]

        
average = round(total/count,2) # calculating the average Cost and round it to 2 decimal places

# looping again to replace missing cost values with the calculated average
for i in range(len(project_dataframe)):
    
    # If cost is missing, replacing it with the average
    if pd.isna(project_dataframe.loc[i,"Cost"]):
        project_dataframe.loc[i,"Cost"] = average

# writing updated dataframe back to the csv file
project_dataframe.to_csv("project.csv",index=False) # Setting index = False so that the DataFrame’s index is not written as an extra column in the CSV file.
```

## Task 3
#### Split the name column in the Employee dataframe into two new columns “First Name” and “Last Name” and remove the older “name” column.

```python
# loading the employee.csv file and loading its contents into the dataframe
employee_dataframe = pd.read_csv("employee.csv") 

# splitting the names and loading it in new columns, first name and last name
employee_dataframe[["First Name","Last Name"]] = employee_dataframe["Name"].str.split(' ',expand = True) 

# deleting the old "Name" column from the original dataframe as instructed
employee_dataframe.drop(["Name"], axis=1, inplace = True) 

# saving the updated dataframe back to csv file
# index=False prevents the dataframe index from being written as a column
employee_dataframe.to_csv("employee.csv", index=False)
```

## Task 4
#### Join all three dataframes in one single dataframe. Name it “Final”

```python
#loading csv files into the respective dataframes
project_dataframe = pd.read_csv("project.csv")
employee_dataframe = pd.read_csv("employee.csv")
seniority_level_dataframe = pd.read_csv("seniority.csv")

#joning the three dataframes into one as "Final"
Final = employee_dataframe.merge(project_dataframe, on="ID", how="inner").merge(seniority_level_dataframe, on="ID", how="inner")

Final
```

## Task 5
#### Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.

```python
# assigning bonus based on project completion status
# if the project Status is 'Finished', assigning 5% of cost as bonus otherwise assigning 0.0
# also adding new column named Bonus

Final["Bonus"] = np.where(Final["Status"] == 'Finished',Final["Cost"]*0.05,0.0) 
```

## Task 6
#### Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.

```python
# demoting the designation level by 1 for employees whose project Status is 'Failed'
Final.loc[Final["Status"]=="Failed", "Designation Level"] += 1

# removing the employee record whose designation level is above 4
Final = Final.loc[Final["Designation Level"] <= 4]
```

## Task 7
#### Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# adding "Mr." to the first name of male employees
Final.loc[Final["Gender"]=="M","First Name"] = "Mr. " + Final.loc[Final["Gender"] == "M", "First Name"]

# adding "Mrs. to the first name of female employees
Final.loc[Final["Gender"]=="F","First Name"] = "Mrs. " + Final.loc[Final["Gender"] == "F", "First Name"]

# deleting the gender column
Final.drop(["Gender"], axis = 1, inplace = True)
```

## Task 8
#### Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.

```python
# updating  the 'Designation Level' column based on age using IF condition as instructed

Final["Designation Level"] = Final.apply(
    lambda x: x["Designation Level"] - 1      # promoting designation level by 1
    if x["Age"] > 29                          # if age greater than 29, promoting
    else x["Designation Level"],              # else keeping the same value
    axis=1                                   # applying the function row by row
)

Final
```

## Task 9
#### Add the cost of all projects for each Employee and save it in new dataframe “TotalProjCost” with three columns ID, First Name, and Total cost

```python
# grouping the dataframe by id and first name
# index = false makes sure that grouped columns remain as regular columns or else both columns would become multi-index
grouped = Final.groupby(["ID","First Name"], as_index = False)

TotalProjCost = grouped["Cost"].sum() #finding the sum of cost by grouped values for each ID–First Name combination

#renaming the columns as instructed 
TotalProjCost.columns = ['ID', 'First Name', 'Total Cost']

TotalProjCost
```

## Task 10
#### Print all the employee details whose city name contains the letter “o” in it.

```python
# filtering final dataframe such that only the employee details 
# whose 'City' name contains the letter 'o' exist.
employee_details = Final[Final["City"].str.contains("o", case = False)]

# displaying the filtered dataframe as instructed
print(employee_details)
```
